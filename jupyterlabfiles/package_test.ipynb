```python
import time

class Timer:    
    def __enter__(self):
        self.start = time.time()
        return self

    def __exit__(self, *args):
        self.end = time.time()
        self.interval = self.end - self.start
```

```python
import numpy as np

import operator
from functools import reduce

file=open('datafiles/pumpkin.txt')    
dataMat=[]  
labelMat=[]
for line in file.readlines():    
    curLine=line.strip().split(" ") 
    floatLine=list(map(float,curLine[1:4]))
    dataMat.append(floatLine) 
faces_pk = reduce(operator.add, dataMat)

file=open('datafiles/pumpkin_v.txt')    
dataMat=[]  
labelMat=[]
for line in file.readlines():    
    curLine=line.strip().split(" ") 
    floatLine=list(map(float,curLine[1:4]))
    dataMat.append(floatLine) 
vertices_pk = reduce(operator.add, dataMat)

faces_num = int(len(faces_pk)/3)
colors_pk = np.linspace(0.5, 0.8, num=faces_num)
```

```python
%%time
dv.passdata_ele(vertices_pk, faces_pk, colors_pk)
```

```python
%%time
dv.visualization_ele(vertices_pk, faces_pk, colors_pk)
```

```python
# Help deal with txt files to be data format that visualization method accepts
def cut(word):
    res = ""
    for i in range(0,word.__len__(),1):
        if word[i] == '/':
            return res
        else:
            res += word[i]
    return res

def getfacetxt(filein, fileout):
    f = open(filein,"r")
    fw = open(fileout,"w")
    lines = f.readlines()

    for i in range(0,lines.__len__(),1):
        output = ""
        for word in lines[i].split():
            word=word.strip('')
            word = cut(word)
            output += word + ' '
        fw.write(output+"\n")

# Get faces and vertices value from .txt files
def getfaces(file):
    file=open(file)    
    dataMat=[]  
    labelMat=[]
    for line in file.readlines():    
        curLine=line.strip().split(" ") 
        floatLine=list(map(float,curLine[1:4]))
        dataMat.append(floatLine) 
    faces_cur = reduce(operator.add, dataMat)
    return faces_cur

def getVer(file):
    file=open(file)    
    dataMat=[]  
    labelMat=[]
    for line in file.readlines():    
        curLine=line.strip().split(" ") 
        floatLine=list(map(float,curLine[1:4]))
        dataMat.append(floatLine) 
    vertices_cur = reduce(operator.add, dataMat)
    return vertices_cur 
```

### Visualization of a boat: vertices(11451), faces(18687), colors(6229)

```python
def cutb(word):
	res = ""
	for i in range(0,word.__len__(),1):
		if word[i] == '/':
			return res
		else:
			res += word[i]
	return res

def getfacetxtb(filein, fileout):

	f = open(filein,"r")
	fw = open(fileout,"w")
	lines = f.readlines()

	for i in range(0,lines.__len__(),1):
		output = ""
		for word in lines[i].split():
			word = word.strip('')
			word = cutb(word)
			output += word + ' '
		if output != "":
			if output[0] != "f":
				output = ""
		if output != "":
			fw.write(output+"\n")

getfacetxtb("datafiles/boat.txt","datafiles/boat_f.txt")
faces = getfaces("datafiles/boat_f.txt")
vertices = getVer("datafiles/boat_v.txt")
```

```python
%%time
dv.passdata_ele_nocolor(vertices, faces)
```

```python
%%time
with Timer() as t:
    dv.visualization_ele(vertices, faces)
print("Visualization time (Boat colored by elements) (vertices(11451), faces(18687), colors(6229))): {0}".format(t.interval))
```
